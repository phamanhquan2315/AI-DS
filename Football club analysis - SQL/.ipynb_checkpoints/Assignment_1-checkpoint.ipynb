## 12. Câu hỏi 12: Liệt kê tên các đội bóng và tổng số bàn thắng mỗi đội, sắp xếp theo số lượng giảm dần

#### tổng số bàn thắng của 1 đội tính cả bàn thắng sân nhà và sân khách
#### sử dụng union

```python
table = pd.read_sql("""select team_long_name, sum(goal) from 
                        (select t.team_long_name,sum(m.home_team_goal) as goal
                        from team t join match m on t.team_api_id = m.home_team_api_id
                        group by t.team_api_id
                        union all
                        select t2.team_long_name,sum(m2.away_team_goal) as goal
                        from team as t2 join match as m2 on t2.team_api_id = m2.away_team_api_id
                        group by t2.team_api_id)
                        group by team_long_name
                        order by goal desc""",conn)
table
```

## 13. Câu hỏi 13: Liệt kê tên các đội bóng (long name) và tổng số trận đấu đội đó tham gia, sắp xếp theo thứ tự giảm dần

### tham gia san nha va san khach

```python
table = pd.read_sql("""select team_long_name, sum(count_match) as sum_match from 
                        (select t1.team_long_name, count(match_api_id) as count_match 
                        from match m1 join team t1 on m1.home_team_api_id = t1.team_api_id
                        group by t1.team_api_id
                        union all
                        select t2.team_long_name, count(match_api_id) as count_match
                        from match m2 join team t2 on m2.home_team_api_id = t2.team_api_id
                        group by t2.team_api_id
                        )
                        group by team_long_name
                        order by sum_match desc
                        """,conn)
table
```

## 14. Câu hỏi 14: Liệt kê số trận thắng, thua và hòa của mỗi đội.

### Nếu một đội có số lượng bàn thắng trên sân nhà > số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thắng" trong trận đấu này
### Nếu một đội có số lượng bàn thắng trên sân nhà < số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thua" trong trận đấu này
### Nếu một đội có số lượng bàn thắng trên sân nhà = số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "hòa" trong trận đấu này
Sử dụng câu lệnh "Case When"
### 1 đội đá cả sân nhà và sân khách

```python
table = pd.read_sql(""" select team_long_name, sum(win),sum(draw),sum(lose) from 
                        (select team_long_name,sum(case when home_team_goal > away_team_goal then 1 else 0 end) as win,
                        sum(case when home_team_goal = away_team_goal then 1 else 0 end) as draw,
                        sum(case when home_team_goal < away_team_goal then 1 else 0 end) as lose
                        from 
                        team t join match m on team_api_id = m.home_team_api_id
                        group by team_long_name
                        
                        union all
                        
                        select team_long_name,sum(case when away_team_goal > home_team_goal then 1 else 0 end) as win,
                        sum(case when away_team_goal = home_team_goal then 1 else 0 end) as draw,
                        sum(case when away_team_goal < home_team_goal then 1 else 0 end) as lose
                        from 
                        team join match on team_api_id = match.away_team_api_id
                        group by team_long_name
                        )
                        group by team_long_name
                        """,conn)
table
```

##  15. Câu hỏi 15: Liệt kê 10 đội có số trận thắng nhiều nhất

```python
table = pd.read_sql("""select team_long_name, sum(win) from
                        (select team_long_name, sum(case when home_team_goal > away_team_goal then 1 else 0 end) as win,
                        sum(case when home_team_goal = away_team_goal then 1 else 0 end) as draw,
                        sum(case when home_team_goal < away_team_goal then 1 else 0 end) as lose
                        from team join match m on team_api_id = m.home_team_api_id
                        group by team_long_name
                        
                        union all
                        select team_long_name, sum(case when away_team_goal > home_team_goal then 1 else 0 end) as win,
                        sum(case when away_team_goal = home_team_goal then 1 else 0 end) as draw,
                        sum(case when away_team_goal < home_team_goal then 1 else 0 end) as lose
                        from team join match m1 on team_api_id = m1.away_team_api_id
                        group by team_long_name
                        )
                        group by team_long_name
                        order by sum(win) desc
                        limit 10""",conn)
table
```

## 16. Câu hỏi 16: Liệt kê tỉ lệ phần trăm về số trận của mỗi giải trên tổng tất cả các trận

```python
table = pd.read_sql("""select * from match""",conn)
table
```

```python
table = pd.read_sql(""" with match_league as (select l.name,count(match_api_id) as count_match from league l join match m on m.league_id = l.id 
                        group by league_id) 
                        select name, count_match*100.0/(select sum(count_match) from match_league) as percent from match_league
                        group by name
                        """,conn)
table
```

## 17. Câu hỏi 17: Liệt kê tỉ lệ phần trăm về số bàn ghi được của từng giải so với tổng số bàn của tất cả các trận

```python
table = pd.read_sql("""with goal_league as (select l.name, sum(home_team_goal + away_team_goal) as sum_goal 
                        from league l join match m on l.id = m.league_id 
                        group by league_id)
                        select name, sum_goal*100.0/(select sum(sum_goal) from goal_league) as percent from goal_league
                        group by name
                    """,conn)
table
```
